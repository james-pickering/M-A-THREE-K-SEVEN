```python
import numpy as np

def bracelet(int1, int2):
    terms = [int1, int2]
    for i in range(2,100):
        temp = (int1 + int2) % 10
        int1 = int2
        int2 = temp
        if (int1 == terms[0] and int2 == terms[1]):
            terms.pop(-1)
            break
        else:
            terms.append(int2)
    return np.array(terms)

for i in range(0,10):
    for j in range(0,10):
        print("(",i,j,")", bracelet(i,j).size)
```
