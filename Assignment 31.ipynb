```python
from random import sample
import matplotlib.pyplot as plt
from scipy.stats import expon
import seaborn as sns
import numpy as np
from collections import Counter


def simulate_paper_draw(n):
    papers = list(range(1, n + 1))
    
    while len(papers) != 1:
        a, b = sample(papers, 2)
        papers.remove(a)
        papers.remove(b)
        difference = abs(b - a)
        papers.append(difference)
        
    return papers[0]
 
simulations = 10**4

for i in range(4,20):

    results = [simulate_paper_draw(i) for _ in range(simulations)]
    element_counts = Counter(results)
    print("Simulation for",i,":")
    for element, count in sorted(element_counts.items()):
        print(f'{element}: {count}')
    print("-------")
```

```python

fig, axes = plt.subplots(2, 4, figsize=(20, 10))


simulations = 10**4  # Reduced number of simulations


for i, ax in enumerate(axes.flatten(), start=4):
    results = [simulate_paper_draw(i) for _ in range(simulations)]
    element_counts = Counter(results)


    sorted_counts = sorted(element_counts.items())
    

    elements, counts = zip(*sorted_counts)
    

    ax.bar(elements, counts, color='black')
    ax.set_title(f'Distribution for N={i}')
    ax.set_xlabel('Final Number')
    ax.set_ylabel('Frequency')
    ax.grid(True)


    ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))

plt.savefig('distribution_plots.jpg', format='jpg', dpi=300)
    
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
