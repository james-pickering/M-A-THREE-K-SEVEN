```python
import matplotlib.pyplot as plt

def P(n, memo):
    if n in memo:
        return memo[n]
    if n == 1:
        return 1
    else:
        memo[n] = 1 - (0.5 * P(n-1, memo))
        return memo[n]

memo = {}

probabilities = {n: P(n, memo) for n in range(1, 26)}

plt.figure(figsize=(10, 6))
plt.plot(list(probabilities.keys()), list(probabilities.values()), marker='o', linestyle='-', color='black')
plt.xlabel('$n$')
plt.ylabel('$P(n)$')
plt.grid(True)
plt.savefig('convergence.jpg', format='jpg', dpi=300)
plt.show()
```
