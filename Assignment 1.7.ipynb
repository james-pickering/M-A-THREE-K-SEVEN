```python
import numpy as np

def main(k):

    def bracelet(int1, int2):
        count = 0
        i1 = int1
        j1 = int2
        for i in range(2,10000):
            temp = (int1 + int2) % k
            int1 = int2
            int2 = temp
            if (int1 == i1 and int2 == j1):
                count += 1
                break
            else:
                count += 1
        return count
    
    maxCount = 0
    
    for i in range(0,k):
        for j in range(0,k):
            if bracelet(i,j) > maxCount:
                maxCount = bracelet(i,j)
    
    return maxCount

for i in range(1,101):
    print(main(i))
```
